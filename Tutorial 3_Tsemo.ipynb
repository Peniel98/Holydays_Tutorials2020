## $\mt{BELL}$ states
The  $\mt{BELL}$ states are very important in quantum information. All the  $\mt{BELL}$ states for a basis called $\mt{BELL}$ basis and all $\ket{\psi}=\alpha\ket{00}+\beta\ket{01}+\gamma\ket{10}+\delta\ket{11} \in \mt{H}^2$ can be expressed in the $\mt{BELL}$ basis like it follows:
$\ket{\psi}=\proj{\mt{B_{00}}}\ket{\psi}+\proj{\mt{B_{01}}}\ket{\psi}+\proj{\mt{B_{10}}}\ket{\psi}+\proj{\mt{B_{11}}}\ket{\psi}=
\fr{1}{\sq{2}}[(\alpha+\delta)\ket{\mt{B_{00}}}+(\alpha-\delta)\ket{\mt{B_{10}}}+(\beta+\gamma)\ket{\mt{B_{01}}}+(\beta-\gamma)\ket{\mt{B_{11}}}]$.

Generally, the $\mt{BELL}$ states can be written like :$$\ket{\mt{B}_{xy}}=\fr{1}{2}(\ket{oy}+(-1)^x\ket{1(1-y)}), x,y\in\{0,1\}$$ where $x$ is the phase bit and $y$ is the control bit.

The $\mt{BELL}$ states respect the following relations:
$$\mt{X}\ot\mt{X}\ket{\mt{B}_{xy}}=(-1)^x\ket{\mt{B}_{xy}}$$
$$\mt{Z}\ot\mt{Z}\ket{\mt{B}_{xy}}=(-1)^y\ket{\mt{B}_{xy}}$$
$$\mt{Y}\ot\mt{Y}\ket{\mt{B}_{xy}}=(-1)^{x+y}\ket{\mt{B}_{xy}}$$

#### Exercise 4.2.2- Density matrix for a quantum ensemble

Given a quantum ensemble prepared with $75\%$ in the state $\ket{a}=\frac{1}{\sqrt{3}}\ket{+}+\sqrt{\frac{2}{3}}\ket{-}$ and $25\%$ in the state $\ket{b}=\frac{2}{3}\ket{+}-\frac{\sqrt{5}}{3}\ket{-}$

1. Let us compute the statistic operator of the quatum ensemble.

This statistic operator is given by $\mc{\rho}=\sum_i\mc{P_i}\ket{\psi_i}\bra{\psi_i}=\sum_i\mc{P_i}\rho_i$ where $\mc{P_i}$ and $\mc{\rho_i}$ are respectively the probabilities and the statistic operators of each single system. In our case the states $\ket{\psi_i}$ are $\ket{a}$ and $\ket{b}$. For the computation, we will call _a_ and _b_ the states $\ket{a}$ and $\ket{b}$.also, the states $\ket{+}$ and $\ket{-}$ will be called $\ket{0}$ and $\ket{1}$.

```python
a=(1/sqrt(3))*(ket("0"))+(sqrt(2)/sqrt(3))*ket("1")
b=(2/3)*(ket("0"))-(sqrt(5)/3)*ket("1")
```

So the density matrix is given by $\rho=\frac{3}{4}\ket{a}\bra{a}+\frac{1}{4}\ket{b}\bra{b}$.Let us call it _rho_ for the computation.

```python
rho=(3/4)*a*a.dag()+(1/4)*b*b.dag()
print("The density matrix is given by ",rho)
```

2.Probabilities to find $\ket{+}$ and $\ket{-}$ if a measurement is made.

By the law of $\mt{Born}$: $\mc{P(a_n})=\Tr({\mc{P_n}\rho})$ where $\mc{P_n}=\ket{n}\bra{n}$.

The probability to find the state $\ket{+}$ is given by $\mc{P_+}=\Tr{(\proj{+}\rho})$ 

```python
P1=((ket("0").proj())*rho).tr()
print('The probability to finding ket(0) is', P1)
```

The probability to find the state $\ket{-}$ is given by $\mc{P_-}=\Tr{(\proj{-}\rho})$ 

```python
P2=((ket("1").proj())*rho).tr()
print('The probability to finding ket(1) is', P2)
```

## Partial trace
The partial ttarce plays an importaznt role in quantum physics because it allows to determine the properties of only one part of coupled system composed of two or several sub-systems. Taking the partial trace of a global system reduces the dimension of the Hilbert space by averaging the global density operator of the system.
In QuTiP, we use the class methos `qutip.Qobj.ptrace` that acts on `qutip.Qobj` instances for wihich it is called.
it takes one argument which is the system competent we are interested by, the other components are traced out.
let take the density operator of a two-coupled system $\ket{\psi}=\ket{01}$, $\rho=\ket{\psi}\bra{\psi}$

```python
psi=tensor(basis(2,0),basis(2,1))
```

```python
psi.ptrace(0) # tracing the second qubit
```

```python
psi.ptrace(1) # tracing the first qubit
```

We can do the same thing when we are dealing with two or several density matrices

```python
rho=tensor(ket2dm(basis(2,0)), fock_dm(2,1))
```

```python
rho.ptrace(0)
```

```python
rho.ptrace(1)
```
